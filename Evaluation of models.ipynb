Visualisation

```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt

# Extracting the last training accuracy
last_resnet_training_acc = resnet_training_acc[-1]
last_mobilenet_training_acc = mobilenet_training_acc[-1]
last_CNN_training_acc = CNN_training_acc[-1]

# Data for the models
data = {
    'Model': ['ResNet50', 'MobileNetv3', 'EfficientNet'],
    'Training Accuracy': [last_resnet_training_acc, last_mobilenet_training_acc, last_CNN_training_acc],
    'Test Accuracy': [resnet_test_acc, mobilenet_test_accuracy, CNN_results]
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame in a tabulated format
print(tabulate(df, headers='keys', tablefmt='pretty'))
```

```python
# Generate x-axis values (epochs or steps)
epochs = range(1, len(resnet_training_loss) + 1)

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot training loss
plt.plot(epochs, resnet_training_loss, marker='o', linestyle='-', color='b', label='ResNet50')

# Plot additional data
plt.plot(epochs, mobilenet_training_loss, marker='s', linestyle='--', color='r', label='MobileNetv3')


# Plot additional data
plt.plot(epochs, CNN_training_loss, marker='^', linestyle=':', color='g', label='CNN')

plt.title('Training Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Generate x-axis values (epochs or steps)
epochs = range(1, len(resnet_training_acc) + 1)

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot training accuracy
plt.plot(epochs, resnet_training_acc, marker='o', linestyle='-', color='b', label='ResNet50')

# Plot additional data
plt.plot(epochs, mobilenet_training_acc, marker='s', linestyle='--', color='r', label='MobileNetv3')

# Plot additional data
plt.plot(epochs, CNN_training_acc, marker='^', linestyle=':', color='g', label='CNN')

plt.title('Training Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.xticks(epochs)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Define the models and their accuracies
models = ['ResNet50', 'MobileNetv3', 'CNN']
test_accuracies = [resnet_test_acc, mobilenet_test_accuracy, CNN_results]

# Create the line plot
plt.figure(figsize=(8, 6))
plt.plot(models, test_accuracies, marker='o', linestyle='-', color='blue', markersize=10)

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy Comparison of Three Models')
plt.ylim(0.75, 1.05)  # Adjust the y-axis limits to focus on the range of values
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Define the models and their accuracies
models = ['ResNet50', 'MobileNetv3', 'CNN']
auc_score = [ResNet_auc_score, mobilenet_auc_score, CNN_auc_score]

# Create the line plot
plt.figure(figsize=(8, 6))
plt.plot(models, auc_score, marker='o', linestyle='-', color='blue', markersize=10)

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('AUC score')
plt.title('AUC score Comparison of Three Models')
plt.ylim(0.8, 1.05)  # Adjust the y-axis limits to focus on the range of values
plt.grid(True)

# Display the plot
plt.show()
```
